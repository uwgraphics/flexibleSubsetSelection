```python
df = pd.read_csv("../data/exampleDatasets/clothing.csv", usecols=range(1,20))

df.dropna(inplace=True)

# convert category label names to integer codes, save codes to dictionaries
df["brand"] = df["brand"].astype("category")
brandCategories = list(df["brand"].cat.categories)
brandCategories = dict(zip(range(0, len(brandCategories)), brandCategories))
df["brand"] = df["brand"].cat.codes

df["ageGroup"] = df["ageGroup"].astype("category")
ageGroupCategories = list(df["ageGroup"].cat.categories)
ageGroupCategories = dict(zip(range(0, len(ageGroupCategories)), ageGroupCategories))
df["ageGroup"] = df["ageGroup"].cat.codes

df["gender"] = df["gender"].astype("category")
genderCategories = list(df["gender"].cat.categories)
genderCategories = dict(zip(range(0, len(genderCategories)), genderCategories))
df["gender"] = df["gender"].cat.codes

df["color"] = df["color"].astype("category")
colorCategories = list(df["color"].cat.categories)
colorCategories = dict(zip(range(0, len(colorCategories)), colorCategories))
df["color"] = df["color"].cat.codes

df["fashionType"] = df["fashionType"].astype("category")
fashionTypeCategories = list(df["fashionType"].cat.categories)
fashionTypeCategories = dict(zip(range(0, len(fashionTypeCategories)), fashionTypeCategories))
df["fashionType"] = df["fashionType"].cat.codes

df["season"] = df["season"].astype("category")
seasonCategories = list(df["season"].cat.categories)
seasonCategories = dict(zip(range(0, len(seasonCategories)), seasonCategories))
df["season"] = df["season"].cat.codes

df["usage"] = df["usage"].astype("category")
usageCategories = list(df["usage"].cat.categories)
usageCategories = dict(zip(range(0, len(usageCategories)), usageCategories))
df["usage"] = df["usage"].cat.codes

df["product"] = df["product"].astype("category")
productCategories = list(df["product"].cat.categories)
productCategories = dict(zip(range(0, len(productCategories)), productCategories))
df["product"] = df["product"].cat.codes

df["category"] = df["category"].astype("category")
categories = list(df["category"].cat.categories)
categories = dict(zip(range(0, len(categories)), categories))
df["category"] = df["category"].cat.codes

df["subcategory"] = df["subcategory"].astype("category")
subcategories = list(df["subcategory"].cat.categories)
subcategories = dict(zip(range(0, len(subcategories)), subcategories))
df["subcategory"] = df["subcategory"].cat.codes

df = df.astype({"returnable": int,
        "exchangeable": int,
        "pickup": int,
        "tryAndBuy": int})

df.set_index("id", inplace=True)
categoricals = { "ageGroup": ageGroupCategories, "usage": usageCategories,
        "gender": genderCategories, "color": colorCategories,
        "fashionType": fashionTypeCategories, "brand": brandCategories,
        "season": seasonCategories, "subcategory": subcategories,
        "product": productCategories, "category": categories}
```
